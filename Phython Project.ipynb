```python
## 3. The following table shows the mean and variance of the expenditure for two groups of people.You want to compare the variability in expenditure
## for both groups with respect to their mean. Which statistical measure would you use to evaluate the variability in expenditure? 
# Please provide an explanation for your answer.

Mean_GroupA = 50000
std_devGroupA = 125000
Mean_GroupB = 40000
std_devGroupB = 10000

## CV=(standard Deviation/ Mean )×100%

# Calculate standard deviation (square root of variance
std_A = np.sqrt(std_devGroupA)
std_B = np.sqrt(std_devGroupB)

# Calculate coefficient of variation
cv_A = (std_A / Mean_GroupA) * 100
cv_B = (std_B / Mean_GroupB) * 100

# Print results
print(f"Group A - Mean: {Mean_GroupA}, Standard Deviation: {std_devGroupA}, Coefficient of Variation: {cv_A:.2f}%")
print(f"Group B - Mean: {Mean_GroupB}, Standard Deviation: {std_devGroupB}, Coefficient of Variation: {cv_B:.2f}%")
```

```python
## 4. During the survey, the ages of 80 patients infected by COVID and 
##admitted to one of the city hospitals were recorded and the collected data is represented in the less than 
##cumulative frequency distribution table.

# Cumulative frequency distribution data
age_ranges = ['5-15', '15-25', '25-35', '35-45', '45-55', '55-65']
No_paitents = [6, 11, 21, 23, 14, 5]

## a. Which class interval has the highest frequency?
# Calculate frequencies for each class interval
frequencies = [No_paitents[0]]  # Frequency for the first interval

for i in range(1, len(No_paitents)):
    frequency = No_paitents[i] - No_paitents[i - 1]
    frequencies.append(frequency)

# Find the index of the interval with the maximum frequency
max_frequency_index = frequencies.index(max(frequencies))

# Retrieve the class interval with the highest frequency
class_interval_with_max_frequency = age_ranges[max_frequency_index]

print(f"The class interval with the highest frequency is: {class_interval_with_max_frequency}")
```

```python
## b. Which age was affected the least?
# Find the index of the interval with the minimum frequency
min_frequency_index = frequencies.index(min(frequencies))

# Retrieve the age range with the least frequency
age_range_least_affected = age_ranges[min_frequency_index]

print(f"The age range affected the least is: {age_range_least_affected}")
```

```python
## c. How many patients aged 45 years and above were admitted?

# Determine how many patients aged 45 years and above
patients_45_and_above = sum(frequencies[4:])  # sum frequencies from index 4 onwards (age range 40-50 and above)

print(f"The number of patients aged 45 years and above admitted is: {patients_45_and_above}")
```

```python
## d. Which is the modal class interval in the above dataset

# Find the index of the interval with the maximum frequency
modal_class_index = frequencies.index(max(frequencies))

# Retrieve the modal class interval
modal_class_interval = age_ranges[modal_class_index]

print(f"The modal class interval in the dataset is: {modal_class_interval}")
```

```python
## e. What is the median class interval of age?
# Total number of observations (in this case, 80 patients)
total_observations = No_paitents[-1]

# Calculate the median position (middle observation)
median_position = (total_observations + 1) / 2

# Find the class interval containing the median
median_class_interval = None
cumulative_sum = 0

for i in range(len(age_ranges)):
    cumulative_sum = No_paitents[i]
    if cumulative_sum >= median_position:
        median_class_interval = age_ranges[i]
        break

# Print the result
print(f"The median class interval of age is: {median_class_interval}")
```

```python
##17. Give the IQR of all the variables which are quantitative and continuous.
# Select quantitative and continuous variables
quantitative_vars = df.select_dtypes(include=['int64', 'float64']).columns

# Calculate IQR for each variable
iqr_values = {}
for var in quantitative_vars:
    q25 = df[var].quantile(0.25)
    q75 = df[var].quantile(0.75)
    iqr = q75 - q25
    iqr_values[var] = iqr

# Print IQR values
print("Interquartile Range (IQR) for quantitative and continuous variables:")
for var, iqr in iqr_values.items():
    print(f"{var}: {iqr:.2f}")
```

```python
## 20. Using the z-score of the income variable, find out the number of observations outside the +-3σ.

# Calculate mean and standard deviation of Income variable
mean_income = df['Income'].mean()
std_dev_income = df['Income'].std()

# Calculate z-scores for Income variable
df['Income_zscore'] = (df['Income'] - mean_income) / std_dev_income

# Count observations outside +-3σ
num_outliers = df[(df['Income_zscore'] > 3) | (df['Income_zscore'] < -3)].shape[0]

print(f"Number of observations outside +-3σ: {num_outliers}")
```
